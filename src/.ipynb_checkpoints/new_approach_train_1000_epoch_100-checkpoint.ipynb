```python
import mnist_loader
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
import new_approach
import time
import numpy as np
```

```python
net1 = new_approach.Network([784, 30, 10], cost=new_approach.CrossEntropyCost)
# net.large_weight_initializer()

t1 = time.time()
eval_cost1, eval_acc1, train_cost1, train_acc1 = \
net1.SGD(training_data[:1000], 100, 50, 1.0, dropout=False, lmbda = 0.0, 
evaluation_data=test_data, monitor_evaluation_cost=False, monitor_evaluation_accuracy=True,
monitor_training_cost=False, monitor_training_accuracy=True)
print time.time() - t1
```

```python
net2 = new_approach.Network([784, 60, 10], cost=new_approach.CrossEntropyCost)
# net.large_weight_initializer()

t1 = time.time()
eval_cost2, eval_acc2, train_cost2, train_acc2 = \
net2.SGD(training_data[:1000], 100, 50, 1.0, dropout=True, lmbda = 0.0, 
evaluation_data=test_data, monitor_evaluation_cost=False, monitor_evaluation_accuracy=True,
monitor_training_cost=False, monitor_training_accuracy=True)
print time.time() - t1
```

```python
import matplotlib.pyplot as plt
plt.plot(range(100), eval_acc1, "b", label="no dropout (600 sec)")
plt.plot(range(100), eval_acc2, "r", label="random dropout (907 sec)")
plt.title("Dropout effect on test accuracy (train size=50000, test size=10000, epoch=100 \
          batch size=50 learning rate=1.0)")
plt.xlabel("epoch")
plt.ylabel("test accuracy")
plt.legend(loc=4)
plt.show()
```

```python
netTest = new_approach.Network([784, 60, 10], cost=new_approach.CrossEntropyCost)
```

```python
netTest.dropVector = [np.random.binomial(1, 0.5, size=hiddenLayerSize) for hiddenLayerSize in netTest.sizes[1:-1]]
```

```python
netTest.dropVector
```

```python
x = np.random.rand(784, 1)
y = np.random.rand(10, 1)
```

```python
zs = []
activations =[x]
def sigmoid(z):
    '''
    The sigmoid function.
    '''
    return 1.0 / (1.0 + np.exp(-z))

sigmoid_vec = np.vectorize(sigmoid)

def sigmoid_prime(z):
    '''
    Derivative of the sigmoid function.
    '''
    return sigmoid(z) * (1 - sigmoid(z))

sigmoid_prime_vec = np.vectorize(sigmoid_prime)
```
