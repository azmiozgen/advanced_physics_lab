```python
import mnist_loader
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
import new_approach
import time
import numpy as np
```

```python
net1 = new_approach.Network([784, 30, 10], cost=new_approach.CrossEntropyCost)
# net.large_weight_initializer()

t1 = time.time()
eval_cost1, eval_acc1, train_cost1, train_acc1 = \
net1.SGD(training_data[:1000], 500, 50, 0.5, dropout=False, lmbda = 0.0, 
evaluation_data=test_data, monitor_evaluation_cost=False, monitor_evaluation_accuracy=True,
monitor_training_cost=False, monitor_training_accuracy=True)
print time.time() - t1
```

```python
net2 = new_approach.Network([784, 60, 10], cost=new_approach.CrossEntropyCost)
# net.large_weight_initializer()

t1 = time.time()
eval_cost2, eval_acc2, train_cost2, train_acc2 = \
net2.SGD(training_data[:1000], 500, 50, 0.5, dropout=True, lmbda = 0.0, 
evaluation_data=test_data, monitor_evaluation_cost=False, monitor_evaluation_accuracy=True,
monitor_training_cost=False, monitor_training_accuracy=True)
print time.time() - t1
```

```python
import matplotlib.pyplot as plt
plt.plot(range(500), eval_acc1, "b", label="no dropout (9907 sec)")
plt.plot(range(500), eval_acc2, "r", label="random dropout (14380 sec)")
plt.title("Dropout effect on test accuracy (train size=1,000, test size=10,000, \n \
epoch=500, batch size=50, eta=0.5)")
plt.xlabel("epoch")
plt.ylabel("test accuracy")
plt.legend(loc=4)
plt.show()
```
